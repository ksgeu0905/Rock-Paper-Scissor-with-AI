```python
import cvzone
import mediapipe
import cv2 as cv
from cvzone.HandTrackingModule import HandDetector
import time, random
```

```python
cap = cv.VideoCapture(0)
cap.set(3,640)
cap.set(4,480)

detector = HandDetector(maxHands=1)


timer = 0
stateResults = False
startGame =False
scores = [0,0] # [AI,Player]
while True:
    imgBG = cv.imread("Resources/Background.png")
    success , img = cap.read()
    
    imgscaled = cv.resize(img,(0,0),None,0.875,0.875)
    imgscaled = imgscaled[:,80:480]
    
    #find hands
    hands, img = detector.findHands(imgscaled)
    
    if startGame:
        
        if stateResults == False:
            timer = time.time() - initialTime
            cv.putText(imgBG,str(int(timer)),(605,435),cv.FONT_HERSHEY_PLAIN,6,(255,0,255),4)

            
        if timer > 3:
            stateResults = True
            timer = 0
            if hands:
                playerMove = None
                hand = hands[0]
                fingers = detector.fingersUp(hand)
                
                if fingers == [0,0,0,0,0]:
                    playerMove = 1
                if fingers == [1,1,1,1,1]:
                    playerMove = 2
                if fingers == [0,1,1,0,0]:
                    playerMove = 3
                
                randomnumber = random.randint(1,3)
                if randomnumber == 1:
                    x="Stone"
                    
                elif randomnumber ==2:
                    x = "Paper"
                    
                else:
                    x = "Scissor"
                
                
                imgAI = cv.imread(f"Resources/{x}.png",cv.IMREAD_UNCHANGED)
                imgBG = cvzone.overlayPNG(imgBG,imgAI,(149,310))
                
                
                
                #player wins
                if (playerMove ==1 and randomnumber== 3) or \
                    (playerMove == 2 and randomnumber ==1) or \
                    (playerMove == 3 and randomnumber==2):
                    scores[1]+=1
                
                    
                #AI wins
                if (playerMove ==3 and randomnumber== 1) or \
                    (playerMove == 1 and randomnumber ==2) or \
                    (playerMove == 2 and randomnumber==3):
                    scores[0]+=1
                    
                if(scores[0]==3):
                    won = "AI WON"
                if(scores[1]==3):
                    won = "YOU WON"
                
                
                print(playerMove)
    
    
    
    
    
    imgBG[234:654,795:1195] = imgscaled
    if stateResults:
        imgBG = cvzone.overlayPNG(imgBG,imgAI,(149,310))
        
        if(scores[0]==3 or scores[1]==3):
            flag = 1
            scores[0]=0
            scores[1]=0
        if flag ==1:
            if won=="YOU WON":
                cv.putText(imgBG,won,(495,175),cv.FONT_HERSHEY_PLAIN,4,(255,0,255),6)
            elif won=="AI WON":
                cv.putText(imgBG,won,(530,175),cv.FONT_HERSHEY_PLAIN,4,(255,0,255),6)
                
            
    cv.putText(imgBG,str(scores[0]),(410,215),cv.FONT_HERSHEY_PLAIN,4,(255,255,255),6)   
    cv.putText(imgBG,str(scores[1]),(1112,215),cv.FONT_HERSHEY_PLAIN,4,(255,255,255),6)   
    
    cv.imshow("Background",imgBG)
    key = cv.waitKey(1)
    if key == ord('s'):
        stateResults = False
        startGame=True
        initialTime = time.time()
        flag=0
        
    if key ==27:
        break
    
cap.release()        
cv.destroyAllWindows()
```
